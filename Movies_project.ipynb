**Categorize the Vote_Average column**

we would cut the Vote_Average  value and make 4 categories: 'Popular' , 'average' , 'below_avg', 'not_popular' to describe it more using 'categorize_col()' function provide

```python
df.head()
```

```python
def categorize_col(df, col, labels):
    if col not in df.columns:
        raise ValueError(f"{col} not found in DataFrame")

    if not pd.api.types.is_numeric_dtype(df[col]):
        raise TypeError(f"{col} must be numeric before categorization")

    if len(labels) != 4:
        raise ValueError("Exactly 4 labels required")

    df[col] = pd.qcut(
        df[col],
        q=4,
        labels=labels,
        duplicates='drop'
    )

    return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'Popular']
df = categorize_col(df, 'Vote_Average', labels)
```
